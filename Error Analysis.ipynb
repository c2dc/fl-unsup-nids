```python
import pandas as pd
pd.set_option('display.max_rows', 500)
```

```python
toniot_preds = pd.read_csv("./extra/toniot_sampled_predicts.csv")
toniot_class = pd.read_csv("./extra/toniot_sampled_test_classes.csv")
nb15_preds = pd.read_csv("./extra/nb15_sampled_predicts.csv")
nb15_class = pd.read_csv("./extra/nb15_sampled_test_classes.csv")

toniot_red_preds = pd.read_csv("./extra/toniot_reduced_predicts.csv")
toniot_red_class = pd.read_csv("./extra/toniot_reduced_test_classes.csv")
nb15_red_preds = pd.read_csv("./extra/nb15_reduced_predicts.csv")
nb15_red_class = pd.read_csv("./extra/nb15_reduced_test_classes.csv")

toniot_sampled = pd.read_csv("./sampled_datasets/toniot_sampled.csv.gz")
toniot_reduced = pd.read_csv("./reduced_datasets/toniot_reduced.csv.gz")
nb15_sampled = pd.read_csv("./sampled_datasets/nb15_sampled.csv.gz")
nb15_reduced = pd.read_csv("./reduced_datasets/nb15_reduced.csv.gz")
```

## Sampled TON-IoT

```python
toniot_sampled_errors = toniot_sampled.iloc[toniot_class["Unnamed: 0"].iloc[toniot_preds["Predict"][toniot_preds.Predict == False].index].values]
```

## Reduced TON-IoT

```python
toniot_reduced_errors = toniot_reduced.iloc[toniot_red_class["Unnamed: 0"].iloc[toniot_red_preds["Predict"][toniot_red_preds.Predict == False].index].values]
```
